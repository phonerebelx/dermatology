```python
import pandas as pd,scipy,numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python

with open('dermatology.names') as fp:
    cols = []
    for line in fp:
        cols.append(line)

new_list = []
for i in range(len(cols)):
    new_list.append(cols[i].replace('\n', ''))

hash_list = []
for i in range(len(new_list)):
    hash_list.append(new_list[i].split())

var = []
for i in range(len(hash_list)):
    for j in range(len(hash_list[i])):
        if "1:" == hash_list[i][j]:
            var.append(hash_list.index(hash_list[i]))
        if "33:" == hash_list[i][j]:
            var.append(hash_list.index(hash_list[i]))
cl2 = []
for i in range(len(hash_list)):
    for j in range(len(hash_list[i])):
        if "Class:" == hash_list[i][j]:

            cl2.append(hash_list.index(hash_list[i]))

inp={}
for i in cl2:
    inp[str(i)] = hash_list[cl2[0]]

for i in inp.values():
    for j in i:
        if "Class" == j:
            i.clear()
            i.append(j)

inp["112:"]=inp[str(cl2[0])]
del inp[str(cl2[0])]

vari = []
hash_list = hash_list[var[0]:var[-1] + 1]
dict_lst = []

for key, value in inp.items():
    dict_lst.append(key)
    dict_lst.append(value[0])
hash_list.append(dict_lst)

imp = 0
while imp < len(hash_list)-1:
    if len(hash_list[imp]) == 0:
        hash_list.remove(hash_list[imp])
        imp -= 1
    elif not hash_list[imp][0][0].isdigit():
        hash_list.remove(hash_list[imp])
    imp+=1




for i in range(len(hash_list)):
  for j in range(len(hash_list[i])):
    vari+=hash_list[i][j].split(":")

lst = []
index = []
shk = ""


for i in range(len(vari)):
  if vari[i].isdigit():
    index.append(int(vari[i]))
    k = i+1
    while k <= len(vari)-1:
      if not vari[k].isdigit():
        shk += vari[k]
        shk += " "

        k += 1
      else:
        lst.append(shk)
        shk = ""
        break
lst.append(shk)


for i in range(len(lst)):

    for j in range(0, len(lst) - i - 1):
        if index[j] > index[j + 1]:
            temp2 = index[j]
            index[j] = index[j + 1]
            index[j + 1] = temp2
            temp = lst[j]
            lst[j] = lst[j + 1]
            lst[j + 1] = temp
print(lst)


```

```python
df = pd.read_csv("dermatology.data")
df = pd.DataFrame(df)
df.columns = lst
df
```

```python
dff = df.iloc[:,-2]
lst = []
for i in dff:
  lst.append(i)

for i in range(len(lst)):
  if lst[i] == "?":
    lst[i] = np.nan
n_arr = np.array(lst)
n_arr = n_arr.astype('float64')
n_arr = n_arr.reshape(-1,1)
n_arr
```

```python
from sklearn.impute import SimpleImputer
imp1 = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
n_arr = n_arr.reshape(-1,1)
n_arr = imp1.fit_transform(n_arr)
n_arr = n_arr.reshape(-1)
n_arr = n_arr.astype("int")
n_arr
```

```python
df.iloc[:,-2] = n_arr
df
```

```python
X = df.iloc[:,:-1]
X
Y = df.iloc[:,-1]
Y
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, Y, test_size=0.2, random_state=42)
X_train_2, X_test_2, y_train_2, y_test_2 =  train_test_split(X, Y, test_size=0.3, random_state=42)
X_train_3, X_test_3, y_train_3, y_test_3 =  train_test_split(X, Y, test_size=0.4, random_state=42)
```

```python

model = GaussianNB()
model.fit(X_train,y_train)
model2 = GaussianNB()
model2.fit(X_train_2,y_train_2)
model3 = GaussianNB()
model3.fit(X_train_3,y_train_3)
```

```python
model.score(X_test,y_test)
model2.score(X_test_2,y_test_2)
model3.score(X_test_3,y_test_3)
predicted = model.predict(X_test_3)
predicted

```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
result = confusion_matrix(y_test_3,predicted)

result
```
